```python
import csv 
```

```python
manufacturers = [] 
```

```python
 with open('car_list.csv', 'rt') as f:
        reader = csv.reader(f)
        for row in reader:
            manufacturers.append(row[0])
```

```python
manufacturers  = manufacturers[1:]
manufacturers
```

```python
set_manunfacturers = set(manufacturers)
```

```python
for manufacturer in set_manunfacturers:
    outfile = open('./' + str(manufacturer) + '.csv', 'w')
    writer = csv.writer(outfile)
    writer.writerow(' ')
    
```
