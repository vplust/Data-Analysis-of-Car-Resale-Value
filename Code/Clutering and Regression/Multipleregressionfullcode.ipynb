```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
import csv 
from statistics import mean 
import glob2 
from sklearn.cluster import KMeans
manufacturers = [] 
new_list = []
with open('car_list.csv', 'rt') as f:
    reader = csv.reader(f)
    for row in reader:
        manufacturers.append(row[0])
manufacturers  = manufacturers[1:]
manufacturers_len = len(set(manufacturers))
print(manufacturers_len)
set_manunfacturers = set(manufacturers)
set_manunfacturers
import pandas as pd
import statsmodels.formula.api as sm
```

```python
cars  = []
with open('car_list.csv', 'rt') as f:
    reader = csv.reader(f)
    for row in reader:
        cars.append(row[1])
cars = cars[1:]
import re
count = 0
new_list = []

for m in set_manunfacturers:
    for car in cars:
        if str(m) in str(car):
            count = count+1
            small_list = []
            new_list.append((re.sub(str(m),"",str(car),count=1)).lower())
car_list = []

for list in new_list:
    car_list.append((' '.join(list.split(' ')[1:-1])).capitalize())
path = '/home/saksham/Desktop/majorproject/out1multipleregression/*.csv'       
print(path)
outfile = open('finalfileregression.csv', 'w')
writer = csv.writer(outfile)
writer.writerow(['MANUFACTURERE', 'CAR', 'intercept', 'year_coeff', 'km_coeff'])
print('len cars' + str(len(car_list)))
new_car_list = []
```

```python
for file in glob2.glob(path):
    print(file)
    for car in car_list:
        
        with open(file, 'rt') as f:
           # x is year
            x = []
            # y is price
            y = []
            # z is kms
            z = []
            reader = csv.reader(f)
            for row in reader:
                if row[1].lower() == car.lower():
                    if car.lower() not  in new_car_list:
                        new_car_list.append(car.lower())
                    c = car
                   
                    x.append( int(2017 - int(row[3])))
                    y.append(int(row[2]))
                    z.append(int(row[4]))
            if len(x)>0:
                man = str(file).split('/')[-1]
                print(man)
                man1 = man.split('.')[0]
                df = pd.DataFrame({"A": y, "B": x, "C": z})
                result = sm.ols(formula="A ~ B + C", data=df).fit()
                print (result.params)
                intercept = result.params[0]
                year_coeff = result.params[1]
                km_coeff = result.params[2]
                outfile = open('finalfileregression.csv', 'a')
                writer = csv.writer(outfile)
                writer.writerow([man1,car,intercept,year_coeff,km_coeff])
```
