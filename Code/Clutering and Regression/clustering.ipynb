```python
from sklearn.cluster import KMeans
x = []
y = []
```

```python
with open('out1/Honda.csv', 'rt') as f:
    reader = csv.reader(f)
    for row in reader:
        if row[1] == 'City':
            x.append( int(2017 - int(row[3])))
            y.append(int(row[2]))
            
```

```python

# solve for a and b
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum(xi*yi for xi,yi in zip(X, Y)) - n * xbar * ybar
    denum = sum(xi**2 for xi in X) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} +  {:.2f}x'.format(a, b))

    return a, b

# solution
a, b = best_fit(xs, ys)
```

```python
plt.scatter(xs, ys)
yfit = [a + b * xi for xi in xs]
plt.plot(xs, yfit)
plt.show()
```
