```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
import csv 
from statistics import mean 
```

```python
from sklearn.cluster import KMeans
# x is year
x = []
# y is price
y = []
# z is kms
z = []
```

```python
with open('out1multipleregression/Honda.csv', 'rt') as f:
    reader = csv.reader(f)
    for row in reader:
        if row[1] == 'City':
            x.append( int(2017 - int(row[3])))
            y.append(int(row[2]))
            z.append(int(row[4]))
```

```python
z
```

```python
import pandas as pd
```

```python
import statsmodels.formula.api as sm
```

```python
df = pd.DataFrame({"A": y, "B": x, "C": z})
```

```python
result = sm.ols(formula="A ~ B + C", data=df).fit()
```

```python
print(result.params)
print(result.params[0])
print(result.params[1])
print(result.params[2])
```

```python
print(result.summary)
```
